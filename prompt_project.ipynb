```python
from transformers import pipeline

# Load the Question Answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
# Define a new context based on the Amazon Rainforest
context = """
The Amazon Rainforest, also known as the Amazon Jungle, is a vast tropical rainforest that spans across nine countries in South America. It is the largest rainforest in the world, covering approximately 5.5 million square kilometers. The Amazon is home to an incredibly diverse range of flora and fauna, with millions of species of plants, animals, and insects. The rainforest plays a crucial role in regulating the Earth's climate, absorbing carbon dioxide and producing oxygen. It is also a source of various medicinal plants and resources for indigenous communities. The Amazon is often referred to as the "lungs of the Earth" due to its vital role in global carbon and oxygen cycles.
"""
question = "What role does the Amazon Rainforest play in regulating the Earth's climate?"

# Get the answer
answer = qa_pipeline(question=question, context=context)

# Print the answer
print(f"Answer: {answer['answer']}")
```
